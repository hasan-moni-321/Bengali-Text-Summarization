```python
import numpy as np 
import pandas as pd 

import os

```

```python
path_text = "/home/hasan/Data Set/Bengali text summarization/Documents"
path_summary = "/home/hasan/Data Set/Bengali text summarization/Summaries"
d = {"Title":[], "Text":[], "Summary1":[], "Summary2":[], "Summary3":[]}
```

```python
for i in range(1, 101):
    doc = os.path.join(path_text, "Document_"+str(i) + ".txt")
    file = open(doc, 'r',  encoding='utf-8-sig')
    file = file.read()
    file = file.replace('Title: ', '')
    file = file.replace('Text: ', '')
    title = file.split('\n')[0]
    if title=='':
        title = file.split('\n')[1]
    if title=='':
        title = file.split('\n')[2]
    d["Title"].append(title)

    del_title = file.replace(title, '')
    text =  del_title.split('\n')
    lines = [line for line in text if line.strip()]
    lines = " ".join(lines)
    d['Text'].append(lines)
    
```

```python
for i in range(1, 101):
    summary1_path = os.path.join(path_summary, "Document_"+str(i)+"_"+ "Summary_1" + ".txt")
    summary2_path = os.path.join(path_summary, "Document_"+str(i)+"_"+ "Summary_2" + ".txt")
    summary3_path = os.path.join(path_summary, "Document_"+str(i) + "_" + "Summary_3" + ".txt")
    
    text1 = open(summary1_path, 'r', encoding='utf-8-sig')
    text1 = text1.read()
    text2 = open(summary2_path, 'r', encoding='utf-8-sig')
    text2 = text2.read()
    text3 = open(summary3_path, 'r', encoding='utf-8-sig')
    text3 = text3.read()
    
    d['Summary1'].append(text1)
    d['Summary2'].append(text2)
    d['Summary3'].append(text3)
```

```python
df = pd.DataFrame(d)
df
```
